```python
import vk_api
import json

with open('../secrets.json', 'r') as f:
    secrets = json.load(f)
```

```python
user_id = secrets['vk']['test_user_id']
vk_session = vk_api.VkApi(
    secrets['vk']['phone'], 
    secrets['vk']['pass'],
    token=secrets['vk']['service_key'],
    app_id=6287487,
    scope=vk_api.VkUserPermissions.GROUPS
)
vk_session.auth(token_only=True)
vk = vk_session.get_api()
```

```python
tools = vk_api.VkTools(vk_session)
```

# Получение статуса

```python
vk_session.method("status.get", {"user_id": user_id})
```

# Получение стены пользователя

```python
vk_session.method("users.get", {
    "user_ids": user_id, 
    'fields': 'activities, about, city, music, universities, career, education, interests,personal,'
})[0]
```
